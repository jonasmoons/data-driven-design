```python
import pandas as pd
```

```python
df = pd.read_json("muacirclejerk.json")
```

```python
sorted_scores = df.sort_values(by='score', ascending=False)
```

```python
sorted_scores.head()
```

```python
sorted_scores["score"].head()
```

```python
sorted_scores.sort_index()
```

```python
sorted_scores[["score", "gilded"]].plot(kind = 'bar')
```

```python
sorted_scores["score"].plot(kind = 'hist', bins = 10)
```

```python
sorted_scores[0:5].plot(kind='barh',x='title',y='score')
```

```python
topscore = sorted_scores["score"][0]
allscore = sorted_scores["score"].sum() - topscore
piechart = pd.DataFrame({
    'Score' : [allscore, topscore]},
    index=['alls','tops'])

piechart
```

```python
piechart.plot(kind='pie',y='Score')
```

```python
sorted_scores[0:5].plot(kind='barh',x='title',y='score', subplots=True)
```

```python
testnew = pd.DataFrame({
    'comments' : sorted_scores["comments"],
    'score' : sorted_scores["score"]})

testnew[0:5].plot(kind='barh')
```
