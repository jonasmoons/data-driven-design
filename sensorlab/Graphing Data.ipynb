```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('jahou-bejeweled.csv')
df = df.drop(["sensor_2"], axis=1)
```

```python
df.head(5)
```

```python
sns.lineplot(y=df['sensor_1'], x=df['timestamp'], data=df)

plt.xlabel('timestamp')
sns.set(rc={'figure.figsize':(11.7,8.27)})

sns.lineplot(y=df['sensor_3'], x=df['timestamp'], data=df)
plt.title('heartrate and pressure while playing Bejeweled')
plt.xlabel('timestamp')
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.show()
```
