```python
#First we're getting the page HTML. This website isn't too strict about accessing its html so we can just import it.
import requests
import pandas as pd

req = requests.get("https://www.twitchmetrics.net/channels/popularity")
html = req.text
```

```python
#importing BeautifulSoup and making sure I'm selecting the right piece of HTML with test=soup.select
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "lxml")

test = soup.select(".list-group-item")
listy = []
```

```python
#With this for loop I'm looping through my earlier generated test. Then I'm appending the info I need to Listy.
#I want the name of the streamer, the language the streamer streams in, the average viewers and the link to the user page.
for names in test:
    listy.append({
    "name" : names.select("h5.mr-2.mb-0")[0].get_text(),
    "language" : names.select("div.mr-3")[0].get_text().strip(),
    "avgviewers" : names.select("samp")[0].get_text(),
    "link" : "https://www.twitchmetrics.net/" + names.select("a")[0].get("href")
    })
```

```python
#Then changing it into a DataFrame to get a nice table.
df = pd.DataFrame(listy)
df.to_csv("best_twitch_streamers.csv")

#I wanted to see the entire link, so I've added this line:
pd.options.display.max_colwidth = 100

#Now I want to visualise our dataframe
df
```

# Repeating the same steps we've used before, but now I'm changing the link to one of the detail pages.
# To Practice, I've made a test list. These are all converted to RAW text since they aren't important for our code.

```python
#To solve our 'looping' issue, we'll loop over our list. Every entry in the list we'll get a new request.
#Then we scrape the urls in entry["link"]
#To finish, we add our new values to the list by adding a new list column with listy[i][name].

i = 0
for entry in listy:
    details_req = requests.get(entry["link"])
    details_html = details_req.text
    details_soup = BeautifulSoup(details_html, "lxml")
    det = details_soup.select(".card-body")
    listy[i]["hours live"] =  det[0].select("div.stat-number.font-weight-bold.my-1")[2].get_text().strip()
    listy[i]["peak viewers"] = det[0].select("div.stat-number.font-weight-bold.my-1")[1].get_text().strip()
    i += 1
```

```python
#Now to visualize in a DataFrame
df_details = pd.DataFrame(listy)
df_details
```

```python
#Saving to csv
df_details.to_csv("detailedtwitchstreamers.csv")
```
